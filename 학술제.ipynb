```python
# 업종 통합

import pandas as pd
import numpy as np

def get_large_category(sector):
    """
    세부 업종(HPSN_MCT_ZCD_NM) 문자열을 입력받아
    대분류 문자열을 반환합니다.
    """
    if pd.isna(sector):
        return np.nan 
    
    # 1. 한식
    if '한식-' in sector or sector in ['백반/가정식', '한정식', '기사식당', '한식뷔페', '구내식당/푸드코트']:
        return '한식'
    
    # 2. 카페/디저트
    if sector in ['카페', '커피전문점', '테이크아웃커피', '테마카페', '차', '주스',
                  '베이커리', '아이스크림/빙수', '도너츠', '떡/한과', '마카롱', '와플/크로플', '탕후루']:
        return '카페/디저트'
    
    # 3. 외식 (기타)
    if '일식-' in sector or sector in ['일식당', '양식', '중식당', '동남아/인도음식', '기타세계요리', '스테이크', 
                                   '중식-딤섬/중식만두', '중식-훠궈/마라탕']:
        return '외식(기타)'
    
    # 4. 간편식/분식
    if sector in ['분식', '치킨', '피자', '햄버거', '샌드위치/토스트', '도시락', '꼬치구이']:
        return '간편식/분식'
        
    # 5. 주점/유흥
    if sector in ['호프/맥주', '요리주점', '이자카야', '민속주점', '포장마차', '와인바', 
                  '일반 유흥주점', '룸살롱/단란주점']:
        return '주점/유흥'

    # 6. 소매/식재료
    if sector in ['축산물', '식료품', '농산물', '수산물', '청과물', '미곡상', '건어물', '유제품', '반찬']:
        return '소매/식재료'

    # 7. 기타/제조
    if sector in ['건강식품', '건강원', '인삼제품', '식품 제조', '떡/한과 제조', '담배', '와인샵', '주류']:
        return '기타/제조'

    return '기타(미분류)'

# --- 메인 코드 실행 ---

input_file = 'big_data_with_hdong_final.csv'
output_file = 'big_data_with_large_categories_FINAL.csv'

print("--- 스크립트 실행 시작 ---")

try:
    print(f"'{input_file}' 파일을 읽고 있습니다...")
    df = pd.read_csv(input_file)
    print("파일 읽기 완료.")

    if 'HPSN_MCT_ZCD_NM' not in df.columns:
        # 'HPSN_MCT_ZCD_NM' 컬럼이 없으면 여기서 KeyError를 발생시킴
        raise KeyError
    
    print(f"'HPSN_MCT_ZCD_NM' 컬럼을 대분류로 변환합니다...")
    df['HPSN_MCT_ZCD_NM'] = df['HPSN_MCT_ZCD_NM'].map(get_large_category)
    print("변환 완료.")

    df.to_csv(output_file, index=False, encoding='utf-8-sig')
    
    print(f"\n--- 작업 성공 ---")
    print(f"'{output_file}' 이름으로 파일 저장을 완료했습니다.")
    print("파일 내용을 확인해보세요.")

except FileNotFoundError:
    print(f"\n[오류] '{input_file}' 파일을 찾을 수 없습니다.")
    print("이 스크립트 파일과 원본 파일이 같은 폴더에 있는지 확인해주세요.")
    
except KeyError:
    # [오류 수정] as e: 부분을 제거하고 가장 단순한 구문으로 변경했습니다.
    print(f"\n[오류] 'HPSN_MCT_ZCD_NM' 컬럼을 찾지 못했습니다.")
    print("원본 파일에 해당 컬럼명이 정확히 존재하는지 확인해주세요.")
    
except Exception as e:
    print(f"\n[알 수 없는 오류] 작업 중 오류가 발생했습니다: {e}")

print("--- 스크립트 실행 종료 ---")
```

# 개월수 별 모델 비교

데이터는 24개월의 데이터를 주며 다른 기준을 정해주지 않았다. 이에 우리는 통상적 경보를 잘 적용할 수 있는 개월수인 2-5개월의 데이터를 각자 최적의 하이퍼파라미터로 모델링 후 학습-평가를 진행하였을 때 가장 좋은 평가의 개월수 모델을 선정하여 진행하기로 했다.

# 불친절 데이터 정제

대회에서 제공받은 데이터에는 23-24년 사이의 정상가게 데이터(24개월)와 폐업가게 데이터(1~24개월 : 폐업 이후에는 데이터가 없음)의 데이터의 개월수가 불규칙적으로 섞여있다. 
따라서 5개월 기준 적은 데이터는 남은 값들의 평균값으로 대체 후 5개월을 채웠고 5개월을 넘는 데이터는 최근 5개월 이후 데이터는 제외하여 데이터프레임을 만들었다.

```python
import pandas as pd
import numpy as np

# 1️⃣ 수치형 컬럼만 추출
numeric_cols = final_df.select_dtypes(include=[np.number]).columns.tolist()

# 2️⃣ '기준년월'을 정수형으로 변환
final_df['기준년월'] = final_df['기준년월'].astype(int)

# 3️⃣ 점포별 그룹화
grouped = final_df.groupby('가맹점ID')

processed_list = []

for store_id, group in grouped:
    group_sorted = group.sort_values('기준년월', ascending=False).reset_index(drop=True)
    n_months = len(group_sorted)

    # 공통 메타 컬럼 (수치형 제외)
    meta_cols = [col for col in group.columns if col not in numeric_cols]

    # ✅ 5개월 이상 → 최근 5개월만 사용
    if n_months >= 5:
        processed = group_sorted.head(5)

    # ✅ 4개월 → 평균값으로 이전 1개월 생성
    elif n_months == 4:
        processed = group_sorted.copy()

        avg_row = group_sorted[numeric_cols].mean().to_dict()
        new_row = {col: avg_row.get(col, np.nan) for col in numeric_cols}

        # 메타 정보는 최신 행 기준으로 복사
        for col in meta_cols:
            new_row[col] = group_sorted.iloc[0][col]

        # 기준년월은 가장 오래된 달보다 1 작게
        new_row['기준년월'] = group_sorted['기준년월'].min() - 1

        processed = pd.concat([processed, pd.DataFrame([new_row])], ignore_index=True)

    # ✅ 3개월 → 평균값으로 이전 2개월 생성
    elif n_months == 3:
        processed = group_sorted.copy()
        avg_row = group_sorted[numeric_cols].mean().to_dict()
        new_rows = []
        min_month = group_sorted['기준년월'].min()

        for i in range(2):
            new_row = {col: avg_row.get(col, np.nan) for col in numeric_cols}
            for col in meta_cols:
                new_row[col] = group_sorted.iloc[0][col]
            new_row['기준년월'] = min_month - (i + 1)
            new_rows.append(new_row)

        processed = pd.concat([processed, pd.DataFrame(new_rows)], ignore_index=True)

    # ✅ 2개월 → 평균값으로 이전 3개월 생성
    elif n_months == 2:
        processed = group_sorted.copy()
        avg_row = group_sorted[numeric_cols].mean().to_dict()
        new_rows = []
        min_month = group_sorted['기준년월'].min()

        for i in range(3):
            new_row = {col: avg_row.get(col, np.nan) for col in numeric_cols}
            for col in meta_cols:
                new_row[col] = group_sorted.iloc[0][col]
            new_row['기준년월'] = min_month - (i + 1)
            new_rows.append(new_row)

        processed = pd.concat([processed, pd.DataFrame(new_rows)], ignore_index=True)

    # ✅ 1개월 이하 → 제외
    else:
        continue

    processed_list.append(processed)

# 4️⃣ 최종 병합
standardized_df = pd.concat(processed_list, ignore_index=True)

# 5️⃣ 최신순 정렬
standardized_df = standardized_df.sort_values(['가맹점ID', '기준년월'], ascending=[True, False]).reset_index(drop=True)
```

# 모델 복잡도 제어

데이터의 n수가 적기 때문에 변수를 늘려 복잡도를 제어했다.

```python
# 기준 컬럼명 지정
target_col = '동일업종매출비율'  # 기준 지표
change_col = '매출변화율'       # 새로 만들 변화율 컬럼명

# 결과 저장용 리스트
change_rows = []

# 점포별 처리
for store_id, group in standardized_df.groupby('가맹점ID'):
    # 최근 기준년월 순으로 정렬
    group_sorted = group.sort_values('기준년월', ascending=False).reset_index(drop=True)

    # ✅ 5개월 데이터 존재하는 경우만 처리
    if len(group_sorted) == 5:
        try:
            v0 = group_sorted.loc[0, target_col]  # 최근달
            v1 = group_sorted.loc[1, target_col]
            v2 = group_sorted.loc[2, target_col]
            v3 = group_sorted.loc[3, target_col]
            v4 = group_sorted.loc[4, target_col]  # 가장 오래된 달

            # 변화율 계산
            r1 = (v0 - v1) / v1 if v1 != 0 else 0
            r2 = (v1 - v2) / v2 if v2 != 0 else 0
            r3 = (v2 - v3) / v3 if v3 != 0 else 0
            r4 = (v3 - v4) / v4 if v4 != 0 else 0

            # 평균 변화율 계산
            avg_change = (r1 + r2 + r3 + r4) / 4

            # 전체 그룹에 동일한 변화율 추가
            group_sorted[change_col] = avg_change

        except Exception:
            group_sorted[change_col] = 0  # 예외 시 0 처리

        change_rows.append(group_sorted)

# ✅ 결과 병합
final_with_change = pd.concat(change_rows, ignore_index=True)
```

```python
import pandas as pd
import numpy as np

def get_large_category(sector):
    """
    세부 업종(HPSN_MCT_ZCD_NM) 문자열을 입력받아
    대분류 문자열을 반환합니다.
    """
    if pd.isna(sector):
        return np.nan 
    
    # 1. 한식
    if '한식-' in sector or sector in ['백반/가정식', '한정식', '기사식당', '한식뷔페', '구내식당/푸드코트']:
        return '한식'
    
    # 2. 카페/디저트
    if sector in ['카페', '커피전문점', '테이크아웃커피', '테마카페', '차', '주스',
                  '베이커리', '아이스크림/빙수', '도너츠', '떡/한과', '마카롱', '와플/크로플', '탕후루']:
        return '카페/디저트'
    
    # 3. 외식 (기타)
    if '일식-' in sector or sector in ['일식당', '양식', '중식당', '동남아/인도음식', '기타세계요리', '스테이크', 
                                   '중식-딤섬/중식만두', '중식-훠궈/마라탕']:
        return '외식(기타)'
    
    # 4. 간편식/분식
    if sector in ['분식', '치킨', '피자', '햄버거', '샌드위치/토스트', '도시락', '꼬치구이']:
        return '간편식/분식'
        
    # 5. 주점/유흥
    if sector in ['호프/맥주', '요리주점', '이자카야', '민속주점', '포장마차', '와인바', 
                  '일반 유흥주점', '룸살롱/단란주점']:
        return '주점/유흥'

    # 6. 소매/식재료
    if sector in ['축산물', '식료품', '농산물', '수산물', '청과물', '미곡상', '건어물', '유제품', '반찬']:
        return '소매/식재료'

    # 7. 기타/제조
    if sector in ['건강식품', '건강원', '인삼제품', '식품 제조', '떡/한과 제조', '담배', '와인샵', '주류']:
        return '기타/제조'

    return '기타(미분류)'

# --- 메인 코드 실행 ---

input_file = 'big_data_with_hdong_final.csv'
output_file = 'big_data_with_large_categories_FINAL.csv'

print("--- 스크립트 실행 시작 ---")

try:
    print(f"'{input_file}' 파일을 읽고 있습니다...")
    df = pd.read_csv(input_file)
    print("파일 읽기 완료.")

    if 'HPSN_MCT_ZCD_NM' not in df.columns:
        # 'HPSN_MCT_ZCD_NM' 컬럼이 없으면 여기서 KeyError를 발생시킴
        raise KeyError
    
    print(f"'HPSN_MCT_ZCD_NM' 컬럼을 대분류로 변환합니다...")
    df['HPSN_MCT_ZCD_NM'] = df['HPSN_MCT_ZCD_NM'].map(get_large_category)
    print("변환 완료.")

    df.to_csv(output_file, index=False, encoding='utf-8-sig')
    
    print(f"\n--- 작업 성공 ---")
    print(f"'{output_file}' 이름으로 파일 저장을 완료했습니다.")
    print("파일 내용을 확인해보세요.")

except FileNotFoundError:
    print(f"\n[오류] '{input_file}' 파일을 찾을 수 없습니다.")
    print("이 스크립트 파일과 원본 파일이 같은 폴더에 있는지 확인해주세요.")
    
except KeyError:
    # [오류 수정] as e: 부분을 제거하고 가장 단순한 구문으로 변경했습니다.
    print(f"\n[오류] 'HPSN_MCT_ZCD_NM' 컬럼을 찾지 못했습니다.")
    print("원본 파일에 해당 컬럼명이 정확히 존재하는지 확인해주세요.")
    
except Exception as e:
    print(f"\n[알 수 없는 오류] 작업 중 오류가 발생했습니다: {e}")

print("--- 스크립트 실행 종료 ---")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df = df.sort_values(['가맹점ID', '기준년월']).reset_index(drop=True)

num_cols = [
    '동일업종매출비율', '동일업종매출건수비율', '동일업종매출순위비율', '동일상권매출순위비율',
    '동일업종해지비율', '동일상권해지비율', '매출변화율',
    '남성 20대이하 고객 비중', '남성 30대 고객 비중', '남성 40대 고객 비중',
    '남성 50대 고객 비중', '남성 60대이상 고객 비중',
    '여성 20대이하 고객 비중', '여성 30대 고객 비중', '여성 40대 고객 비중',
    '여성 50대 고객 비중', '여성 60대이상 고객 비중',
    '재방문 고객 비중', '신규 고객 비중',
    '거주 이용 고객 비율', '직장 이용 고객 비율', '유동인구 이용 고객 비율',
    '동별_점포수'
]

# -----------------------------
# 변화율, 평균, 표준편차 계산 (index 정렬 안전하게)
# -----------------------------
for col in num_cols:
    grp = df.groupby('가맹점ID')[col]
    df[f'{col}_5개월변화율'] = grp.transform(lambda x: x.pct_change(periods=5))
    df[f'{col}_5개월평균'] = grp.transform(lambda x: x.rolling(window=5, min_periods=1).mean())
    df[f'{col}_5개월표준편차'] = grp.transform(lambda x: x.rolling(window=5, min_periods=1).std())

# -----------------------------
# 기울기(추세) 계산 함수
# -----------------------------
def slope(series):
    """5개월 구간 선형회귀 기울기"""
    if series.isnull().any() or len(series) < 5:
        return np.nan
    x = np.arange(len(series)).reshape(-1, 1)
    y = series.values
    model = LinearRegression().fit(x, y)
    return model.coef_[0]

for col in num_cols:
    df[f'{col}_5개월추세'] = (
        df.groupby('가맹점ID')[col]
        .transform(lambda x: x.rolling(window=5, min_periods=5).apply(slope, raw=False))
    )

# -----------------------------
# 결측/무한값 정리
# -----------------------------
df.replace([np.inf, -np.inf], np.nan, inplace=True)

print(f"✅ 생성된 변화율/추세 피처 수: {len(num_cols) * 4}")
print(f"✅ 전체 컬럼 수: {df.shape[1]}")
```

# 시계열 데이터 처리

시계열 데이터인 위 데이터의 경우 기존의 CatBoost를 시차(lag)·rolling feature 등과 같이 쓰거나 CNN/LSTM으로 빠지는 방법이 있지만 가게별 5개월 데이터를 그룹화하여 하나로 취급 후 샘플링.

```python
# ==========================================
# 📦 라이브러리
# ==========================================
import pandas as pd
import numpy as np
from sklearn.model_selection import GroupKFold
from sklearn.utils import resample
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from catboost import CatBoostClassifier

# ==========================================
# 1️⃣ 언더샘플링 (폐업:비폐업 균형 조정)
# ==========================================
df_minority = df[df['폐업여부'] == 1]
df_majority = df[df['폐업여부'] == 0]

df_majority_down = resample(df_majority,
                            replace=False,
                            n_samples=len(df_minority) * 5,
                            random_state=42)

df_balanced = pd.concat([df_majority_down, df_minority]).sample(frac=1, random_state=42).reset_index(drop=True)

print("언더샘플링 후 데이터 크기:", df_balanced.shape)
print(df_balanced['폐업여부'].value_counts())

# ==========================================
# 2️⃣ 데이터 분리
# ==========================================
X = df_balanced.drop(columns=['가맹점ID', '기준년월', '폐업여부'])
y = df_balanced['폐업여부']
groups = df_balanced['가맹점ID']  # GroupKFold에서 동일 점포 분리 기준

# ==========================================
# 3️⃣ Group K-Fold 설정
# ==========================================
gkf = GroupKFold(n_splits=5)
results = []

# ==========================================
# 4️⃣ 교차검증 루프
# ==========================================
for fold, (train_idx, val_idx) in enumerate(gkf.split(X, y, groups)):
    print(f"\n===== Fold {fold+1} =====")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    # ------------------------------
    # (A) CatBoost 1단계 학습
    # ------------------------------
    model = CatBoostClassifier(
        iterations=1000,
        depth=8,
        learning_rate=0.01,
        l2_leaf_reg=2,
        subsample=0.9,
        rsm=0.9,
        class_weights=[1, 20],  # 폐업 가중치 ↑
        eval_metric='Recall',
        random_seed=42,
        verbose=False
    )
    model.fit(X_train, y_train)

    # ------------------------------
    # (B) 중요 피처 80개만 선별
    # ------------------------------
    fi = pd.DataFrame({
        'feature': X.columns,
        'importance': model.get_feature_importance()
    })
    top_features = fi.sort_values('importance', ascending=False).head(80)['feature'].tolist()

    X_train_sel = X_train[top_features]
    X_val_sel = X_val[top_features]

    # ------------------------------
    # (C) 상위 피처만 재학습
    # ------------------------------
    model.fit(X_train_sel, y_train)

    # ------------------------------
    # (D) 예측 및 평가
    # ------------------------------
    y_proba = model.predict_proba(X_val_sel)[:, 1]
    threshold = 0.25   # recall 중심 threshold
    y_pred = (y_proba >= threshold).astype(int)

    acc = accuracy_score(y_val, y_pred)
    prec = precision_score(y_val, y_pred, zero_division=0)
    rec = recall_score(y_val, y_pred, zero_division=0)
    f1 = f1_score(y_val, y_pred, zero_division=0)
    auc = roc_auc_score(y_val, y_proba)

    results.append([fold + 1, acc, prec, rec, f1, auc])

# ==========================================
# 5️⃣ 교차검증 성능 요약
# ==========================================
cv_results = pd.DataFrame(results, columns=['Fold', 'Accuracy', 'Precision', 'Recall', 'F1', 'ROC-AUC'])

print("\n📊 교차검증별 성능:\n", cv_results)
print("\n✅ 교차검증 평균 성능")
print(f"Accuracy: {cv_results['Accuracy'].mean():.4f}")
print(f"Precision: {cv_results['Precision'].mean():.4f}")
print(f"Recall: {cv_results['Recall'].mean():.4f}")
print(f"F1-score: {cv_results['F1'].mean():.4f}")
print(f"ROC-AUC: {cv_results['ROC-AUC'].mean():.4f}")
```

```python
# ==========================================
# 🧩 전처리 함수 정의 (훈련/예측 공용)
# ==========================================
import pandas as pd
import numpy as np

def preprocess_features(df: pd.DataFrame) -> pd.DataFrame:
    """
    balanced_data.csv에서 사용한 주요 파생 변수 생성 로직을
    store_features.csv에도 동일하게 적용하는 함수
    """
    # ✅ 1. 운영개월수구간 생성
    if "운영개월수" in df.columns and "운영개월수구간" not in df.columns:
        df["운영개월수구간"] = pd.cut(
            df["운영개월수"],
            bins=[0, 12, 24, 36, 48, 60, np.inf],
            labels=[1, 2, 3, 4, 5, 6]
        ).astype(float)

    # ✅ 2. 매출변화율, 매출변화율_3개월평균 등 계산
    if all(col in df.columns for col in ["매출금액_1개월전", "매출금액_2개월전", "매출금액_3개월전"]):
        df["매출변화율"] = (
            (df["매출금액_1개월전"] - df["매출금액_2개월전"]) / (df["매출금액_2개월전"] + 1e-6)
        )
        df["매출변화율_3개월평균"] = (
            (df["매출금액_1개월전"] + df["매출금액_2개월전"] + df["매출금액_3개월전"]) / 3
        ) / (df["매출금액_3개월전"] + 1e-6) - 1

    # ✅ 3. 성별/연령별 고객 비중 이동평균 (없으면 건너뜀)
    for col in df.columns:
        if ("고객 비중" in col) and ("_4개월평균" not in col):
            new_col = f"{col}_4개월평균"
            df[new_col] = (
                df.groupby("가맹점ID")[col].transform(lambda x: x.rolling(4, 1).mean())
                if "가맹점ID" in df.columns else df[col]
            )

    # ✅ 4. 비율형 지표의 결측치 보정
    ratio_cols = [c for c in df.columns if "비율" in c or "순위" in c]
    df[ratio_cols] = df[ratio_cols].fillna(df[ratio_cols].median())

    # ✅ 5. 결측치 전체 보정
    df = df.fillna(0)

    return df
```

# 챗봇 성능 보안

챗봇이 단순히 +- 값만으로 판단하지 않도록 성능을 보안.

```python
# ==========================================
# 0️⃣ 라이브러리
# ==========================================
import pandas as pd
import numpy as np
import joblib
import shap
import google.generativeai as genai


# ==========================================
# 1️⃣ 환경 설정 (API 키, 모델 로드)
# ==========================================
GEMINI_API_KEY = "api"
genai.configure(api_key=GEMINI_API_KEY)
llm = genai.GenerativeModel("models/gemini-2.5-pro")

# CatBoost 모델 로드
model = joblib.load("final_catboost_model.pkl")

# 모델이 학습 시 사용한 정확한 feature 이름 가져오기
trained_features = model.feature_names_



# ==========================================
# 2️⃣ 예측용 데이터 준비
# ==========================================
df = pd.read_csv("store_features.csv")

# 🔹 (여기 추가!) 전처리 함수 적용
df = preprocess_features(df)

# 🔹 모델 피처 순서 맞추기
X = df.reindex(columns=trained_features, fill_value=0)



# ==========================================
# 3️⃣ 예측 및 위험 등급 계산
# ==========================================
pred_prob = model.predict_proba(X)[:, 1][0]

if pred_prob >= 0.30:
    risk_level = "높음"
elif pred_prob >= 0.15:
    risk_level = "중간"
else:
    risk_level = "낮음"


# ==========================================
# 4️⃣ SHAP 영향도 계산
# ==========================================
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)

row_idx = 0
sample_shap = shap_values[row_idx]

# 상위 5개 중요 변수 추출
top_k = 10
feature_importance = sorted(
    zip(trained_features, sample_shap),
    key=lambda x: abs(x[1]),
    reverse=True
)[:top_k]

# 영향도 텍스트 생성
shap_explain_lines = []
for fname, sval in feature_importance:
    direction = "위험↑ 요인" if sval > 0 else "안정↑ 요인"
    shap_explain_lines.append(f"- {fname}: 영향도 {sval:.4f} ({direction})")

feature_text = "\n".join(shap_explain_lines)

print("📊 주요 영향 변수:")
print(feature_text)


# ==========================================
# 5️⃣ 폐업점포 기준 임계값 계산 (개선 버전)
# ==========================================
df_train = pd.read_csv("balanced_data.csv")
df_closed = df_train[df_train["폐업여부"] == 1]

# 상위 피처 추출
top_features = [f[0] for f in feature_importance]

# ✅ 존재하는 컬럼만 선택 (부분 일치 허용)
available_features = []
for f in top_features:
    matches = [col for col in df_closed.columns if f.replace(" ", "") in col.replace(" ", "")]
    if matches:
        available_features.append(matches[0])

# ✅ 임계값 계산
thresholds = df_closed[available_features].agg(["mean", "median"])
thresholds.loc["q25"] = df_closed[available_features].quantile(0.25)
thresholds.loc["q75"] = df_closed[available_features].quantile(0.75)

# ✅ SHAP 방향성 기반 위험 방향 설정
feature_direction = {}
for fname, sval in feature_importance:
    if sval > 0:
        feature_direction[fname] = "high"  # 값이 높을수록 위험
    else:
        feature_direction[fname] = "low"   # 값이 낮을수록 위험

# ✅ 현재 매장 데이터에서 동일 컬럼 추출
current_values = {}
for f in available_features:
    match = [col for col in df.columns if f.replace(" ", "") in col.replace(" ", "")]
    if match:
        current_values[f] = df[match[0]].iloc[0]

# ✅ 비교 및 해석 텍스트 생성
threshold_text = ""
for col in available_features:
    if col not in current_values:
        threshold_text += f"- {col}: ❌ 현재 데이터에 해당 컬럼 없음\n"
        continue

    val = current_values[col]
    avg = thresholds.loc["mean", col]
    q25 = thresholds.loc["q25", col]
    q75 = thresholds.loc["q75", col]
    direction = feature_direction.get(col, "unknown")

    # ⚙️ 위험 판단 로직 (방향성 반영)
    if direction == "high":
        status = "⚠️ 위험구간" if val >= q75 else "✅ 안정범위"
    elif direction == "low":
        status = "⚠️ 위험구간" if val <= q25 else "✅ 안정범위"
    else:
        status = "⚪ 판단불가"

    threshold_text += (
        f"- {col}: 현재={val:.3f}, 폐업점 평균={avg:.3f}, 방향={direction} ({status})\n"
    )

print("\n📏 주요 변수 임계값 비교 (방향성 반영):")
print(threshold_text)
```

```python
# ==========================================
# 6️⃣ Gemini 리포트 생성
# ==========================================
prompt = f"""
당신은 프랜차이즈 본사의 데이터 기반 경영 컨설턴트입니다.

이 가맹점(ID: {store_id})의 폐업 위험 예측 확률은 {pred_prob:.2%} (위험 수준: {risk_level}) 입니다.
모델이 판단한 주요 영향 요인은 다음과 같습니다:

{feature_text}

또한 아래는 폐업 점포들의 평균 임계값과 비교한 현재 매장의 실제 값입니다.
(⚠️은 위험구간, ✅은 안정구간을 의미)

{threshold_text}

이 정보를 바탕으로 다음을 작성하세요:
1. 폐업 위험이 {risk_level}으로 나타난 주요 데이터 기반 이유
2. 임계값을 기준으로 현재 가맹점이 주의해야 할 지표 2가지
3. 3개월 내 개선 가능한 핵심 실천 전략 2가지
4. 같은 동 내 같은 업종 평균 대비 강점과 약점
5. 사장님이 이해하기 쉬운 언어로, 위협적이지 않게 코칭하듯 작성

반드시 한국어로, 친절하고 실용적으로 작성하세요.
"""

response = llm.generate_content(prompt)

print("\n📋 Gemini 리포트 요약\n")
print(response.text)
```
